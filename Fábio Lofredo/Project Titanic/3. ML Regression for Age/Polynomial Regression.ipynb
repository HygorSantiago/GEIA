# Importing data and libraries.

```python
import pandas as pd
import numpy as np
```

```python
data_raw=pd.read_csv('/content/data_num.csv',index_col=0)
data_raw
```

# Cleaning

Some file cleaning has already been done previously

Dropping items that will not be used in the analysis

```python
data = data_raw.copy()
data.drop(["Name","Survived"],inplace=True,axis=1)
data.dropna(inplace=True)
print(data.isna().sum())
print(len(data))
```

Separating variables, independent variables X and dependent variable y for split and train

```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
print(X,y)
```

# PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
# Tested also with degree 1,2,3,4,5
poly_degree = PolynomialFeatures(degree = 2)
X_poly = poly_degree.fit_transform(X)
```

# Split into training & test sets

The data will be separated into four groups

Independent variables:
- X_train: for train the model
- X_test: for prediction

Dependent variables:
- y_train: for train the model
- y_test: for prediction

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size = 0.2, random_state = 0)
```

```python
print('---X_train---'*20)
print(X_train)
print('---X_test---'*20)
print(X_test)
print('---y_train---'*20)
print(y_train)
print('---y_test---'*20)
print(y_test)
```

#Scaling


.fit() Calculate the mean and standard deviation to use on transform(), and store the values in the StandardScaler() class instance, in this case in the variable sc

.transform() perform standardization by the following formula:
$\mathrm{x}_{standardization}^{}=\frac{x-mean}{standard\text{ }deviation}$

----------------------------

print(sc.mean_) to see the mean 

print(sc.var_) to see the variance

print(np.sqrt(sc.var_)) or print(sc.scale_) to see the standard deviation

-----------------------------

The dummy variables aren't standardized because they are already between 0 and 1

----------------------------

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

From https://www.atoti.io/articles/when-to-perform-a-feature-scaling/ about scaling in linear regressions:

"Technically, feature scaling does not make a difference in the regression, but it might give us some practical benefits and in further feature engineering steps."

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,:-10] = sc.fit_transform(X_train[:,:-10])
X_test[:,:-10] = sc.transform(X_test[:,:-10])
```

```python
print('---X_train---'*20)
print(X_train)
print('---X_test---'*20)
print(X_test)
```

# Modelling

## Build the model

Instantiate the object with the name regressor as a model, for the machine learning

**degree** is the degree of the polynomial features

**X_poly** is the number of polynomial features generated from the combination of inputs. So it will be much larger than input.

This **.fit_transform** is different from the sklearn.preprocessing.StandardScaler



Adapted from:

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler.fit_transform


https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html

```python
from sklearn.linear_model import LinearRegression

parameters = {'copy_X': True, 'fit_intercept': True, 'n_jobs': 0, 'positive': False}

#** send the value of dict to the function
regressor = LinearRegression(**parameters)
```

## Train the model

Train the model on classifier object, with X_train and y_train

It is not testing with X_train but with X_poly

```python
regressor.fit(X_train, y_train)
```

## Make predictions

Predictions for dependent variables with the trained model

Prediction for the vector X_test

```python
prediction = regressor.predict(X_test)
#np.set_printoptions(precision=2)
#prediction = regressor.predict(X_test)
print(prediction.mean())
print(y_test.mean())
```

# Evaluation

 R²(coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). In the general case when the true y is non-constant, a constant model that always predicts the average y disregarding the input features would get a R² score of 0.0.

Source: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

```python
from sklearn.metrics import r2_score
r2_score(y_test, prediction)
```
