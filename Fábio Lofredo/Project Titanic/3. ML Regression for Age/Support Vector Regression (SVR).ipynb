# Importing data and libraries.

```python
import pandas as pd
import numpy as np
```

```python
data_raw=pd.read_csv('/content/data_num.csv',index_col=0)
data_raw
```

# Cleaning

Some file cleaning has already been done previously

Dropping items that will not be used in the analysis

```python
data = data_raw.copy()
data.drop(["Name","Survived"],inplace=True,axis=1)
data.dropna(inplace=True)
print(data.isna().sum())
print(len(data))
```

Separating variables, independent variables X and dependent variable y for split and train

```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
print(X,y)
```

Reshape y for scaling

```python
y = y.reshape(len(y),1)
```

#Scaling


.fit() Calculate the mean and standard deviation to use on transform(), and store the values in the StandardScaler() class instance, in this case in the variable sc

.transform() perform standardization by the following formula:
$\mathrm{x}_{standardization}^{}=\frac{x-mean}{standard\text{ }deviation}$

----------------------------

print(sc.mean_) to see the mean 

print(sc.var_) to see the variance

print(np.sqrt(sc.var_)) or print(sc.scale_) to see the standard deviation

-----------------------------

The dummy variables aren't standardized because they are already between 0 and 1

----------------------------

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

In this SVR we will be scaling first, because we will performace the scaling in y too

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X[:,:-10] = sc_X.fit_transform(X[:,:-10])
y = sc_y.fit_transform(y)
```

```python
print('---X_train---'*20)
print(X)
print('---X_test---'*20)
print(y)
```

# Split into training & test sets

The data will be separated into four groups

Independent variables:
- X_train: for train the model
- X_test: for prediction

Dependent variables:
- y_train: for train the model
- y_test: for prediction

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
np.set_printoptions(threshold=200)
print('---X_train---'*20)
print(X_train)
print('---X_test---'*20)
print(X_test)
print('---y_train---'*20)
print(y_train)
print('---y_test---'*20)
print(y_test)
```

# Modelling

## Build the model

Instantiate the object with the name regressor as a model, for the machine learning

```python
from sklearn.svm import SVR
parameters = {'C': 1, 'coef0': 8, 'degree': 2, 'gamma': 'auto', 'kernel': 'poly'}
regressor = SVR(**parameters)
```

## Train the model

Train the model on classifier object, with X_train and y_train

```python
regressor.fit(X_train, y_train)
```

## Make predictions

Predictions for dependent variables with the trained model

Prediction for the vector X_test

```python
prediction  = sc_y.inverse_transform(regressor.predict(X_test).reshape(-1,1))
prediction
```

# Evaluation

 R²(coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). In the general case when the true y is non-constant, a constant model that always predicts the average y disregarding the input features would get a R² score of 0.0.

Source: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

```python
from sklearn.metrics import r2_score
r2_score(sc_y.inverse_transform(y_test), prediction)
```
