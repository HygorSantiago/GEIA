# Importing data and libraries.

```python
import pandas as pd
import numpy as np
```

```python
data_raw=pd.read_csv('/content/data_num.csv',index_col=0)
data_raw
```

# Cleaning

Some file cleaning has already been done previously

Dropping items that will not be used in the analysis

```python
data = data_raw.copy()
data.drop(["Name","Survived"],inplace=True,axis=1)
data.dropna(inplace=True)
print(data.isna().sum())
print(len(data))
```

Separating variables, independent variables X and dependent variable y for split and train

```python
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
print(X,y)
```

# Split into training & test sets

The data will be separated into four groups

Independent variables:
- X_train: for train the model
- X_test: for prediction

Dependent variables:
- y_train: for train the model
- y_test: for prediction

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
np.set_printoptions(threshold=200)
print('---X_train---'*20)
print(X_train)
print('---X_test---'*20)
print(X_test)
print('---y_train---'*20)
print(y_train)
print('---y_test---'*20)
print(y_test)
```

#Scaling


.fit() Calculate the mean and standard deviation to use on transform(), and store the values in the StandardScaler() class instance, in this case in the variable sc

.transform() perform standardization by the following formula:
$\mathrm{x}_{standardization}^{}=\frac{x-mean}{standard\text{ }deviation}$

----------------------------

print(sc.mean_) to see the mean 

print(sc.var_) to see the variance

print(np.sqrt(sc.var_)) or print(sc.scale_) to see the standard deviation

-----------------------------

The dummy variables aren't standardized because they are already between 0 and 1

----------------------------

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

from https://www.atoti.io/articles/when-to-perform-a-feature-scaling/ about using scale for tree based machine learning:

"The decision tree splits a node on a feature that increases the homogeneity of the node. This split on a feature is not influenced by other features. Hence, there is virtually no effect of the remaining features on the split. This is what makes them invariant to the scale of the features."

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,:-10] = sc.fit_transform(X_train[:,:-10])
X_test[:,:-10] = sc.transform(X_test[:,:-10])
```

```python
print('---X_train---'*20)
print(X_train)
print('---X_test---'*20)
print(X_test)
```

# Modelling

## Build the model

Instantiate the object with the name regressor as a model, for the machine learning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import linear_model


parameters = {'max_depth': 6, 'max_features': 'sqrt', 'min_samples_split': 3, 'n_estimators': 150 }

regressor = RandomForestRegressor(**parameters, random_state=0)
```

## Train the model

Train the model on classifier object, with X_train and y_train

```python
regressor.fit(X_train, y_train)
```

## Make predictions

Predictions for dependent variables with the trained model

Prediction for the vector X_test

```python
prediction = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(prediction)
```

# Evaluation

 R²(coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). In the general case when the true y is non-constant, a constant model that always predicts the average y disregarding the input features would get a R² score of 0.0.

Source: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

```python
from sklearn.metrics import r2_score
r2_score(y_test, prediction)
```

# Predicting Age

Checking NaN

```python
data_age = data_raw.copy()
data_age.isna().sum()
```

Predicting and changing the age value in data_age

```python
age = []
import math
for i ,v in data_age.iterrows():
  if math.isnan(v['Age']):
    X_prediction = v[1:-2].values
    X_prediction[:-10] = sc.transform(X_prediction[:-10].reshape(1, -1))
    prediction = regressor.predict(X_prediction.reshape(1, -1))
    data_age.iloc[i, data_age.columns.get_loc('Age')] = prediction[0]
    #print(i, v)
    age.append(prediction[0])
```
