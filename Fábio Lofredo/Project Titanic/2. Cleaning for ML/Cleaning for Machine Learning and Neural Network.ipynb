# Importing data and libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
data_raw_train=pd.read_csv('/content/train.csv')
data_raw_test=pd.read_csv('/content/test.csv')
```

Joining Data

```python
# starting the index of data_raw_2 at the end of the data_raw_1
data_raw_test.index = range(len(data_raw_train),len(data_raw_train)+len(data_raw_test),1)
```

```python
data_raw = pd.concat([data_raw_train,data_raw_test])
data_raw
```

#Changing data to numbers.

## NaN

Checking NaN

```python
data_raw.isna().sum()
```

Copy of original, to keep the original

```python
data_num = data_raw.copy()
```

Replacing NaN

Dropping Ticket and Cabin because they don't have a good correlation

```python
data_num.drop(['Ticket',"Cabin","PassengerId"],axis=1,inplace=True)
```

Replacing NaN in Embarked column with Unknown 

```python
data_num['Embarked'].fillna('Unknown',inplace=True)
```

Replacing 1 NaN value in Fare column with mean 

```python
data_num['Fare'].fillna(data_num['Fare'].mean(),inplace=True)
```

Checking

```python
data_num.isna().sum()
```

## Numbers

Replacing the values in Sex and Embarked to numbers

* Sex

  * male = 0

  * female = 1

* Embarked
  
  * Unknown = 0

  * C = 1

  * Q = 2

  * S = 3

```python
data_num["Sex"].replace(['male','female'],[0,1],inplace=True)
data_num["Embarked"].replace(['Unknown','C','Q','S'],[0,1,2,3],inplace=True)
print(data_num[["Sex","Embarked"]])
```

Creating the column Title, because it could be relevant to analysis

* Tile

  * 2 = "Capt.", "Col.", "Major.", "Dr." or "Rev." (Crew)

  * 3 = "Jonkheer.", "Don.", "Sir.", "the Countess." or "Lady." (Royal)
  
  * 4 = "Mme.", "Ms." or "Mrs." (Married women)

  * 5 = "Mlle." or "Miss." (Single women)

  * 6 = "Master." (Heir)

  * 7 = "Mr." (Men)

  * 0 = Others

```python
#Creating a column with numbers representing the titles

data_num['Title'] = data_num.index*10

print(type(data_num['Name']))
for i , v in data_num['Name'].items():
  if ('Capt. ' in v) or ('Col. ' in v) or ('Major. ' in v) or ('Dr. ' in v) or ('Rev. ' in v):
    data_num['Title'].iloc[i]= 2
  elif ('Jonkheer. ' in v) or ('Don. ' in v) or ('Sir. ' in v) or ('the Countess. ' in v) or ('Lady. ' in v):
    data_num['Title'].iloc[i]= 3
  elif ('Mme. ' in v) or ('Ms. ' in v) or ('Mrs. ' in v):
    data_num['Title'].iloc[i]= 4
  elif ('Mlle. ' in v) or ('Miss. ' in v):
    data_num['Title'].iloc[i]= 5
  elif ('Master. ' in v):
    data_num['Title'].iloc[i]= 6
  elif ('Mr. ' in v):
    data_num['Title'].iloc[i]= 7
  else:
    data_num['Title'].iloc[i]= 0

print(data_num['Title'].value_counts())
print(data_num)
```

## Dummy Variables

Dummy Variables for Embarked

```python
#dividing embarked categorical places into columns
data_num['Embarked Unknown'] = data_num.index*0
data_num['Embarked Cherbourg'] = data_num.index*0
data_num['Embarked Queenstown'] = data_num.index*0
data_num['Embarked Southampton'] = data_num.index*0
embarked = data_num.columns.get_loc("Embarked")
embarked_U = data_num.columns.get_loc("Embarked Unknown")
embarked_C = data_num.columns.get_loc("Embarked Cherbourg")
embarked_Q = data_num.columns.get_loc("Embarked Queenstown")
embarked_S = data_num.columns.get_loc("Embarked Southampton")

for id,row in data_num.iterrows():
  if data_num.iloc[id,embarked] == 0:
    data_num.iloc[id,embarked_U] =1
  elif data_num.iloc[id,embarked] == 1:
    data_num.iloc[id,embarked_C] =1
  elif data_num.iloc[id,embarked] == 2:
    data_num.iloc[id,embarked_Q] =1
  elif data_num.iloc[id,embarked] == 3:
    data_num.iloc[id,embarked_S] =1
  else:
    print("Error")

# Dropping Unknown because the dummy variables need less 1 variable
data_num.drop(["Embarked","Embarked Unknown"], axis=1,inplace=True)

data_num
```

Dummy Variables for Title

```python
#dividing Title categorical places into columns

data_num['Crew'] = data_num.index*0
data_num['Royal'] = data_num.index*0
data_num['Married women'] = data_num.index*0
data_num['Single women'] = data_num.index*0
data_num['Heir'] = data_num.index*0
data_num['Men'] = data_num.index*0
data_num['Others'] = data_num.index*0

Title = data_num.columns.get_loc("Title")
Crew = data_num.columns.get_loc("Crew")
Royal = data_num.columns.get_loc("Royal")
Married_women = data_num.columns.get_loc("Married women")
Single_women = data_num.columns.get_loc("Single women")
Heir = data_num.columns.get_loc("Heir")
Men = data_num.columns.get_loc("Men")
Others = data_num.columns.get_loc("Others")

for id,row in data_num.iterrows():
  if data_num.iloc[id,Title] == 2:
    data_num.iloc[id,Crew] =1
  elif data_num.iloc[id,Title] == 3:
    data_num.iloc[id,Royal] =1
  elif data_num.iloc[id,Title] == 4:
    data_num.iloc[id,Married_women] =1
  elif data_num.iloc[id,Title] == 5:
    data_num.iloc[id,Single_women] =1
  elif data_num.iloc[id,Title] == 6:
    data_num.iloc[id,Heir] =1
  elif data_num.iloc[id,Title] == 7:
    data_num.iloc[id,Men] =1
  elif data_num.iloc[id,Title] == 0:
    data_num.iloc[id,Others] =1
  else:
    print("Error")

# Dropping Others because the dummy variables need less 1 variable
data_num.drop(["Title","Others"], axis=1,inplace=True)

data_num
```

putting the dummy variables, name, and dependent variable at the end

R²(coefficient of determination) regression score function.

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). In the general case when the true y is non-constant, a constant model that always predicts the average y disregarding the input features would get a R² score of 0.0.
